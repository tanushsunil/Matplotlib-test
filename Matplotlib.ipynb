**Matplotlib - Pie Chart, Bar Plot, Sine Wave**

**1.Import** **libraries** 

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

**2. Creating datas**

```python
Subjects = ['tamil','english','mathematics','science','social']
Percentage = [68,70,58,76,85]
```


**scatter plot**


```python
plt.scatter(Subjects,Percentage)
plt.show()
```

**pie** **chart**

```python
plt.pie(Percentage, labels=Subjects)
```

**Bar** **chart**

```python
plt.bar(Subjects, Percentage)
```

**sin** **wave** **plot**

```python
import numpy as np
```

**Range** **function**

```python
time = np.arange(1, 10 , 0.6)
```

**sin** **operation**

```python
amplitude = np.sin(time)
```

```python
plt.plot(time, amplitude)
```

**Contour** **plt**

```python
#creating data
data1 = np.linspace(-6, 6, 100)
data2 = np.linspace(-6, 6, 100)
```

```python
#creating it as 2d
[x, y] = np.meshgrid(data1, data2)
```

```python
# creating 3rd data
z = x**2 + y**2
```

```python
# plotting contour
figure, ax = plt.subplots(1,1)
ax.contourf(x, y, z)
ax.set_title('Contourf Plot(Filled)')
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

**box** **plot**

```python
data1 = np.random.normal(100, 20, 100)
data2 = np.random.normal(70, 40,100)
data3 = np.random.normal(50, 80, 90)
```

```python
data = [data1, data2, data3]
```

```python
figure = plt.figure(figsize=(9,6))
ax = figure.add_axes([0,0,1,1])
ax.boxplot(data)
plt.show()
```

10. 3d **scatter** **plot**

10.1 **importing** **necassary** **libraries**

```python
from mpl_toolkits.mplot3d import Axes3D
import random
```

10.2 **creating** **datas**

```python
list_x = list(range(0, 100))
list_y = list(range(100, 200))
list_z = list(range(200, 300))
```

10.3 **shuffling** **the** **data**

```python
random.shuffle(list_x)
random.shuffle(list_y)
random.shuffle(list_z)
```

10.4. **Plotting** **3D** **scatter**

```python
figure = plt.figure()
ax = Axes3D(figure)

ax.scatter(list_x, list_y, list_z)
plt.show()
```

11. **3D** **Surface** **plot**

11.1. **Creating** **datas**


```python
x = np.outer(np.linspace(-5,5,25), np.ones(25))
y = x.copy().T
z = (np.sin(x**2) + np.cos(y**2))
```

**Plotting 3D surface**

```python
figure = plt.figure(figsize = (10, 7))
ax = plt.axes(projection = '3d')
ax.plot_surface(x, y, z)
plt.show()
```
